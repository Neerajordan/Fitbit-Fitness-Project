```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("/content/minute_stats_data.csv")
```

```python
df.head(4)
```

TOP & BOTTOM USERS BY MET

```python
df= df[df['mets'] != 0]

top = df.loc[df.groupby(['activity_minute'])['mets'].idxmax()].sort_values('mets', ascending=False).head(10)
bottom = df.loc[df.groupby(['activity_minute'])['mets'].idxmin()].sort_values('mets').head(10)


top[['id', 'activity_day', 'activity_minute', 'mets']]
```

```python
bottom[['id', 'activity_day', 'activity_minute', 'mets']]
```

CORREALTION ANALYSIS

```python
crr = df[['mets', 'calories', 'intensity', 'sleep', 'steps']].corr()
crr
```

1. Heat Map

```python
num_column = df.select_dtypes(include='number')
matrix = num_column.corr()

plt.figure(figsize =(8,6))
sns.heatmap(matrix, annot = True, cmap = 'Reds', fmt = '.2f', linewidths = 0.5)
plt.title('Correlation Matrix')
plt.show()
```

2. Scatter Plot

```python
fig1 = px.scatter(df, x='mets', y='calories', title='Relationship between METs and Calories',
                  labels={'metabolism', 'calories'},
                  color_discrete_sequence=['darkviolet'])
fig1.show()
```

3. Pair Plot

```python
fig = px.scatter_matrix(df, dimensions=['calories', 'intensity', 'sleep'],
                        title='PAIR PLOT BETWEEN CALORIES, INTENSITY & SLEEP',
                        labels={'calories': 'Calories', 'intensity': 'Intensity','sleep': 'Sleep'},
                        color_discrete_sequence=['maroon'])

fig.show()
```
